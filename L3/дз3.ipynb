```python
import tensorflow as tf
print(tf.__version__)
msg = tf.constant('TensorFlow 2.0 Hello World')
tf.print(msg)
from __future__ import absolute_import, division, print_function, unicode_literals

# TensorFlow and tf.keras

from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

```

```python
import pandas as pd
```

```python
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images = train_images / 255.0

test_images = test_images / 255.0
```

```python
optimizer_list=['adam', 'RMSprop']
layers_list=[1,2,3]
neurons_list=[128,256]
```

```python
num=0
df=pd.DataFrame(columns=['number','optimizer','layers_number','neurons_number','test_accuracy'])
for optimizer in optimizer_list:
  for layers in layers_list:
    for neurons in neurons_list:
      for _ in range(3):
        model = keras.Sequential()
        model.add(keras.layers.Flatten(input_shape=(28, 28)))
        
        for _ in range(layers):
          n_num=neurons
          model.add(keras.layers.Dense(n_num, activation='relu'))
          n_num=n_num/2
        
        model.add(keras.layers.Dense(10, activation='softmax'))   
        model.compile(optimizer=optimizer,
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
        model.fit(train_images, train_labels, epochs=4, batch_size=100)
        test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
        print('\nTest accuracy:', test_acc)
        df.loc[num, 'number'] = num
        df.loc[num, 'optimizer'] = optimizer
        df.loc[num, 'layers_number'] = layers
        df.loc[num, 'neurons_number'] = neurons
        df.loc[num, 'test_accuracy'] = test_acc


        num+=1
```

```python
df['test_accuracy']=df['test_accuracy'].astype(float)
```

```python
df.dtypes
```

```python
answer=df.groupby(by=['optimizer','layers_number','neurons_number']).median()
answer
```

```python
В целом, увеличение количества нейронов в слое улучшало качество сети, а усложнение сети в виде увеличения количества скрытых слоев с 2 до 3, напротив, ухудшало результаты. 
```
